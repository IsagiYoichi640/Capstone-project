# Data Visualization


*   Get insigts from the data(understand the data)
*   Find the relationships in the data



##1. Pairplot
* Demonstrating the general relationship between data.

```python
sns.pairplot(df)
```

```python
sns.countplot(data=df, x="sex", palette="Purples")
```

### Findings
* There is not significant difference in the number of male and female species because the length of the bar is almost similar.

```python
sns.countplot(data=df, x="island", palette="RdPu");
```

### Findings
* The Biscoe island contains the largest number of penguins because it has the longest bar.
* The Torgersen island contains the smallest number of penguins because it has the shortest bar.

```python
sns.countplot(data=df, x="species", palette="Oranges");
```

### Findings
* The Adelie species is the largest species because it has the longets bar.
* The Chinstrap species are endangered species as they have the shortest bar.

```python
sns.countplot(data=df, x="sex", palette="rocket", hue='species');
```

### Findings
* There is no major difference in the sex counts of all the species since the bar length are similar.

```python
sns.countplot(data=df, x="island", palette="husl", hue="species");
```

### Findings
* The Gento species is only found on Biscoe island where they are the largest count of species on the Island.
* Adelie species are found in all the 3 islands and they are the only species present on Torgersen Island.
* Chinstrap species is only found in Dream Island.
